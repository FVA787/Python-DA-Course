```python
# Если в функции нет return, либо он пустой, то функция возвращает None (специальный тип данных, который означает отсутствие значения). Посмотрим на примере:
def square_3 (number):
    result = number ** 2
    print("hello world")
    print(result*2)
    return  #- когда return без значения, то функция print(ниже) возвратит None. Строчка return должна быть в функции последней, после нее код функции не выполняется. Это то же самое что return None.
    return None
    # return - не обязателен. В этом случае функция вернет None. 
print(square_3(number=5))  # Этой строкой мы выводим на экран результат работы функции. Но поскольку в функции у нас нет return (т.е. функция ничего НЕ ВОЗВРАЩАЕТ), то 
# при команде "выведи результат работы функции на экран" мы видим None. Print(result) - это часть работы функции (то есть функция ВОЗВРАЩАЕТ ДЕЙСТВИЕ вывода на экран сама по себе), но 
# не выдает результат своей работы. Поэтому при запуске выводится 25 (5^2 = 25) как действие функции, но при выводе print() ничего не возвращается, так как нет return.
# print(square_3(number=5))
```

```python
# Функция без параметров. Вводные данные (число) она получает в процессе своей работы, эту функцию нужно просто вызвать. 
def square_2():
    user_input = int(input('введите число'))
    result = user_input ** 2
    return result*3 - 2
square_2()
```

```python
#Функция с двумя параметрами. Все параметры обязательны
def power(number, number_2):
    result = number * number_2
    return result
power(3,2)
```

```python
#Функция с двумя параметрами. Если один из них указать = значение, то этот параметр станет необязательным. 
def power(number, number_2=2):
    result = number * number_2
    return result
power(3)
```

```python
number = 5    #Область видимости. Сначала функция ищет переменную внутри фаункции и, только в случае, если не находит, ищет ее вне функции. 
power = 3
def power_4():
    number = 4
    return number ** power
print(power_4())
print(number ** power)
```

```python
name = "James"     #Мы можем объявить следующим синтаксисом глобальную переменную внутри функции, но делать это крайне не рекомендуется. 
def say_hello():
    global name
    name = "Oleg"
    print("Hello,", name)
say_hello()
print(name)
```

```python
#Лямбда функции - анонимные функции. Объявляются слеюущим синтаксисом. Это удобно с очень простыми функциями:
myfunc = lambda x, y: x + y 
myfunc (10, 5)
```

```python
nums = [1, 2, 3, 4, 5]
odd_even_numbers = lambda x: 'even' if x % 2 == 0 else 'odd'  # пишем слово lambda, затем переменная, затем через двоеточие что хотим получить затем условие (if any - как в list comprehension)
for x in nums:
    print(f'x is {odd_even_numbers(x)}')
```

```python
list(map(odd_even_numbers, nums))
```
