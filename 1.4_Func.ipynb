```python
# Если в функции нет return, либо он пустой, то функция возвращает None (специальный тип данных, который означает отсутствие значения). Посмотрим на примере:
def square_3 (number):
    result = number ** 2
    print("hello world")
    print(result*2)
    return  #- когда return без значения, то функция print(ниже) возвратит None. Строчка return должна быть в функции последней, после нее код функции не выполняется. Это то же самое что return None.
    return None
    # return - не обязателен. В этом случае функция вернет None. 
print(square_3(number=5))  # Этой строкой мы выводим на экран результат работы функции. Но поскольку в функции у нас нет return (т.е. функция ничего НЕ ВОЗВРАЩАЕТ), то 
# при команде "выведи результат работы функции на экран" мы видим None. Print(result) - это часть работы функции (то есть функция ВОЗВРАЩАЕТ ДЕЙСТВИЕ вывода на экран сама по себе), но 
# не выдает результат своей работы. Поэтому при запуске выводится 25 (5^2 = 25) как действие функции, но при выводе print() ничего не возвращается, так как нет return.
# print(square_3(number=5))
```

```python
nums = [1, 2, 3, 4, 5]
odd_even_numbers = lambda x: 'even' if x % 2 == 0 else 'odd'  # пишем слово lambda, затем переменная, затем через двоеточие что хотим получить затем условие (if any - как в list comprehension)
for num in nums:
    print(f'x is {odd_even_numbers(num)}')
```

```python
list(map(odd_even_numbers, nums))   #функция map - предназначена для того, чтобы применить функцию к списку

# Функция map() — это встроенный инструмент Python, который применяет заданную функцию к каждому элементу итерируемого объекта (например, списка или кортежа). 
# Как это работает
# Вы передаете функции map() два основных аргумента:
# Функцию (что нужно сделать).
# Последовательность (с чем это нужно сделать). 
# Она проходит по всем элементам, выполняет над ними действие и возвращает итератор — объект, из которого можно по очереди доставать результаты. 

# Синтаксис
# map(function, iterable) 

# Пример использования
# Допустим, у нас есть список строк, и мы хотим превратить их в числа:

# strings = ['1', '2', '3']
# numbers = map(int, strings)

# # Чтобы увидеть результат в виде списка, нужно использовать list()
# print(list(numbers)) 
# # Результат: [1, 2, 3]
```

```python
list(map(odd_even_numbers, [1])) # код также будет работать
list(map(odd_even_numbers, 1))   # такой код не будет работать, здесь нет списка. 
```

```python
strings = ['1', '2', '3']
list(map(int, strings))

# int — это встроенная функция (как и встроенный тип данных) в Python.
# Ее не нужно объявлять, потому что она доступна «из коробки» сразу после установки языка, наряду с такими функциями как print(), len(), str() или list().
# Почему map(int, strings) работает именно так?
# Когда вы передаете int внутрь map, происходит следующее:
# Функция map берет первый элемент из списка strings — строку '1'.
# Она «скармливает» эту строку функции int, выполняя по сути int('1').
# Полученное целое число 1 сохраняется в итераторе.
# Процесс повторяется для всех элементов списка.
# Что еще можно использовать вместо int?
# В map можно передавать любую функцию, которая принимает один аргумент:
# map(len, ['apple', 'banana']) — посчитает длину каждого слова.
# map(str.upper, ['a', 'b']) — переведет буквы в верхний регистр.
# map(float, ['1.5', '2']) — превратит строки в числа с плавающей точкой.
# Важный нюанс: Обратите внимание, что внутри map мы пишем просто имя функции int, без скобок. Если написать int(), Python попытается выполнить функцию сразу, 
# а нам нужно просто передать «ссылку» на неё, чтобы map использовала её позже.
```

```python
# В map также можно передавать любые списки, в том чилсе range, то есть с генерацией на лету:
def square(x):
    return x ** 2
list(map(square, range(0, 10)))   # мы испльзуем list, так как функция map возвращает объект, который является генератором, то есть он не сразу все объекты выполняет. Функция list их собирает
# в один объет - список.  
```

```python
# *args and **kwargs
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Олег", age=30, city="Москва", children=3)
```

```python
#Почему лямбда функция ? Почему она называется безымянной, хотя, с виду, у нее есть имя? Это имя не функции, это имя переменной, куда мы кладем безымянную функцию. 
# Почему она всё равно анонимная?
# Когда вы пишете def my_func():, Python делает две вещи одновременно:
# Создает объект-функцию.
# Навсегда приклеивает к этому объекту внутренний атрибут __name__, равный 'my_func'.
# Когда же вы пишете odd_even_numbers = lambda x: ..., происходит другое:
# Python создает «безымянный» объект-функцию. Внутри него, в поле __name__, написано просто '<lambda>'.
# Вы сами берете этот безымянный объект и вручную кладете его в переменную odd_even_numbers.
# Как это проверить?
# Попробуйте вывести внутреннее имя функции в консоль:
# Обычная функция

def myfunct(x): return x
print(myfunct.__name__) 
# Выведет: 'standard_func'

# Лямбда-функция
anonymous_func = lambda x: x
print(anonymous_func.__name__) 
# Выведет: '<lambda>'

# В Python встроенные системные свойства всегда начинаются и заканчиваются двумя нижними подчеркиваниями с каждой стороны (так называемые 
# «дандер-методы» — от Double Underscore).
```
