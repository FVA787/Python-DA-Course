```python
# Нужно вывести элементы с конца так, чтобы каждый последующий становился ключом нового словаря, а предыдущий словарь - значением нового словаря.
# Ответ, который нужно получить: {'2023-01-01': {'yandex': {'cpc': 100}}}

some_list = ['2023-01-01', 'yandex', 'cpc', 100]
def get_crazy_nested_list(some_list):
    res = some_list[-1]
    for el in reversed(some_list[:-1]):
        res = {el:res}
    return res
get_crazy_nested_list(some_list)


```

```python
# То же самое можно сделать с помощью функции reduce из пакета functools:

from functools import reduce
some_list = ['2023-01-01', 'yandex', 'cpc', 100]
print(reduce(lambda el1, el2: {el2 : el1}, reversed(some_list)))
```

```python
#Отсортируем список студентов по результатам пяти контрольных работ (вес 0,6) и результату экзамена (вес 0,4), выведем имя, фамилию и показатель. 
# Будем использовать для этого функцию sorted. sorted(iterable, key=None, reverse=False)  — это универсальный инструмент для сортировки данных в 
# Python. Она принимает любой итерируемый объект (список, кортеж, строку, словарь) и всегда возвращает новый отсортированный список.

students_list = [
    {"name": "Василий", "surname": "Теркин", "gender": "м", "program_exp": True, "grade": [8, 8, 9, 10, 10], "exam": 9},
    {"name": "Мария", "surname": "Павлова", "gender": "ж", "program_exp": True, "grade": [7, 8, 9, 7, 9], "exam": 9},
    {"name": "Ирина", "surname": "Андреева", "gender": "ж", "program_exp": False, "grade": [10, 9, 8, 10, 7], "exam": 7},
    {"name": "Татьяна", "surname": "Сидорова", "gender": "ж", "program_exp": False, "grade": [7, 8, 8, 9, 8], "exam": 10},
    {"name": "Иван", "surname": "Васильев", "gender": "м", "program_exp": True, "grade": [9, 8, 9, 6, 9], "exam": 5},
    {"name": "Роман", "surname": "Золотарев", "gender": "м", "program_exp": False, "grade": [8, 9, 9, 6, 9], "exam": 6}
]
def sort_students(students, reverse=True):
    res = sorted(students, reverse = reverse, key = lambda student: sum(student['grade'])/len(student['grade'])*0.6 + student['exam']*0.4)
    for student in res:
        print(f'{student["name"]} {student["surname"]}: {(sum(student["grade"])/len(student["grade"]))*0.6 + (student["exam"])*0.4 :.2f}')
sort_students(students_list)
```

Функции-декораторы

```python
# Timer Decorator. Он позволяет добавить функции новую логику, не меняя её собственный код.
# В основе декораторов лежат те самые функции высшего порядка: одна функция принимает другую, 
# добавляет к ней «декор» (какое-то действие до или после) и возвращает её обновленной.
import time

def timer_decorator(func):
    def wrapper():
        start = time.time()
        func()  # Запускаем основную функцию
        end = time.time()
        print(f"Функция выполнялась {end - start} сек.")
    return wrapper

@timer_decorator            # Это «магия» декоратора
                            # Символ @timer_decorator: Это сокращенная запись. На самом деле Python делает так: 
                            # say_hello = timer_decorator(say_hello). Определяем функцию say_hello() прямо под @time_decorator - такой синтаксис декораторов.
def say_hello():
    time.sleep(1)
    print("Привет!")

say_hello()

# В чем разница между return wrapper и return wrapper()?
# Это самая частая ошибка новичков:
# return wrapper — возвращает саму функцию (её можно запустить позже).
# return wrapper() — сразу запускает функцию и возвращает результат её выполнения (число, строку и т.д.).
# В декораторах нам нужно именно вернуть саму «обертку», чтобы она заменила собой оригинал и ждала вызова.
# Хотите посмотреть, как изменить декоратор, чтобы он мог работать с любым количеством аргументов в функции?



```
