Но эту задачу можно решить в одну строчку кода. Рассмотрим, как: (с помощью list comprehension)

```python
data = [
    [13, 25, 23, 34, 18],
    [45, 34, 44, 47, 22],
    [12, 33, 23, 95, 41],
    [13, 53, 34, 35, 21],
    [34, 45, 14, 91, 12]
]
print([el[index] for index, el in enumerate(data)])
print(sum([el[index] for index, el in enumerate(data)]))
```

```python
people = {
    1: {'имя': 'Олег', 'возраст': '32', 'пол': 'м'},
    2: {'имя': 'Екатерина', 'возраст': '21', 'пол': 'ж'},
    3: {'имя': 'Кира', 'возраст': '24', 'пол': 'м'},
    4: {'имя': 'Владимир', 'возраст': '36', 'пол': 'ж'}
}
sum_ = 0
for person in people.values():
    sum_ += int(person['возраст'])
print((sum_) / len(people))   
```

```python
# Это же можно сделать в одну строку кода, с помощью list comprehension (dict comprehension):

[int(person['возраст']) for _, person in people.items()]  #выполняем соглашение, когда на вход идут две переменные ключ и значение (первую переменную игнорируем, нужно только значение), без вывода для информации.
[int(person['возраст']) for person in people.values()]   # но можно и проще через values, без вывода, для информации.


sum([int(person['возраст']) for person in people.values()]) / len(people)
```

```python
some_text = '''
Python — высокоуровневый язык программирования общего назначения с динамической строгой типизацией и автоматическим управлением памятью, 
ориентированный на повышение производительности разработчика, читаемости кода и его качества, а также на обеспечение переносимости написанных на нём программ. 
Язык является полностью объектно-ориентированным в том плане, что всё является объектами. 
Необычной особенностью языка является выделение блоков кода пробельными отступами.
'''
import re
res = {}
for word in re.sub('[^\w\s]', '', some_text).lower().split():
    if word in res:
        res[word] += 1
    else:
        res[word] = 1
for i, count in res.items():
    print(f'{i}: {count}')

# res = {i for i in re.sub('[^\w\s]', '', some_text).split() : 'A'}
# cleaned_text = re.sub('[^\w\s]', '', some_text).split()


```

```python
#Либо его можно сделать следующим способом, через setdefault:

some_text = '''
Python — высокоуровневый язык программирования общего назначения с динамической строгой типизацией и автоматическим управлением памятью, 
ориентированный на повышение производительности разработчика, читаемости кода и его качества, а также на обеспечение переносимости написанных на нём программ. 
Язык является полностью объектно-ориентированным в том плане, что всё является объектами. 
Необычной особенностью языка является выделение блоков кода пробельными отступами.
'''
import re
res = {}
for word in re.sub('[^\w\s]', '', some_text).lower().split():
    res.setdefault(word, 0)
    res[word] += 1
for i, count in res.items():
    print(f'{i}: {count}')
```
