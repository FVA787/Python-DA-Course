```python
#Либо его можно сделать следующим способом, через setdefault:

some_text = '''
Python — высокоуровневый язык программирования общего назначения с динамической строгой типизацией и автоматическим управлением памятью, 
ориентированный на повышение производительности разработчика, читаемости кода и его качества, а также на обеспечение переносимости написанных на нём программ. 
Язык является полностью объектно-ориентированным в том плане, что всё является объектами. 
Необычной особенностью языка является выделение блоков кода пробельными отступами.
'''
import re
res = {}
for word in re.sub('[^\w\s]', '', some_text).lower().split():
    res.setdefault(word, 0)
    res[word] += 1
for i, count in res.items():
    print(f'{i}: {count}')
```

```python
# Можно решить эту задачу через множества:
some_text = '''
Python — высокоуровневый язык программирования общего назначения с динамической строгой типизацией и автоматическим управлением памятью, 
ориентированный на повышение производительности разработчика, читаемости кода и его качества, а также на обеспечение переносимости написанных на нём программ. 
Язык является полностью объектно-ориентированным в том плане, что всё является объектами. 
Необычной особенностью языка является выделение блоков кода пробельными отступами.
'''
import re
res = {}  #создаем словарь!!!
split_text = re.sub('[^\s\w]', '', some_text).lower().split()

for word in set(split_text):   # создаем множество, так как в множестве нет повторяющихся элементов, не нужно будте делать проверку, и далее  нужно будет добавить только count:
    res[word] = split_text.count(word)  # синтаксис именно таков
for word, count in res.items():
    print(word, count)
```

```python
geo_logs = [
    {'visit1': ['Москва', 'Россия']},
    {'visit2': ['Дели', 'Индия']},
    {'visit3': ['Владимир', 'Россия']},
    {'visit4': ['Лиссабон', 'Португалия']},
    {'visit5': ['Париж', 'Франция']},
    {'visit7': ['Тула', 'Россия']},
    {'visit9': ['Курск', 'Россия']},
    {'visit10': ['Архангельск', 'Россия']}
]

# geo_logs_copy = geo_logs.copy()
# geo_logs_copy = geo_logs[:]
# geo_logs_copy = list(geo_logs)
geo_logs_copy = [
    {'visit3': ['Владимир', 'Россия']},
    {'visit2': ['Дели', 'Индия']},
    {'visit4': ['Лиссабон', 'Португалия']},
    {'visit5': ['Париж', 'Франция']},
    {'visit7': ['Тула', 'Россия']},
    {'visit9': ['Курск', 'Россия']},
    {'visit1': ['Москва', 'Россия']},
    {'visit10': ['Архангельск', 'Россия']}
]

for log in geo_logs:
    if 'Россия' not in list(log.values())[0]:
        geo_logs_copy.remove(log)
print(geo_logs_copy)
print()

# Почему работает этот код? Как он понимает, что log из копии это тот же самый log из оригинального списка словарей? 
# Ответ на этот вопрос подводит нас к разнице между «тем же самым объектом» (identity) и «таким же объектом» (equality).
# Код работает благодаря тому, как устроен метод .remove().
# Секрет метода .remove()
# Когда мы говорим geo_logs_copy.remove(log), Python не ищет в памяти именно тот «физический» объект, который сейчас лежит в переменной log. Вместо этого он:
# Берет содержимое log (например, {'visit2': ['Дели', 'Индия']}).
# Идет в список geo_logs_copy и сравнивает этот словарь с каждым элементом по очереди.
# Для сравнения он использует оператор == (равно по значению).
# В Python два словаря равны, если у них абсолютно одинаковые ключи и значения. Ему неважно, созданы они в разных строчках кода или один является копией другого. 
# Если они «выглядят» одинаково, .remove() считает, что цель найдена, и удаляет элемент.

# Проверим это с помощью следующейго кода:

for log in geo_logs:
    print(log)
    for copy_log in geo_logs_copy:
        if log == copy_log:
             print("Они равны по значению")
        if log is copy_log:
             print("Это один и тот же объект в памяти")
```
