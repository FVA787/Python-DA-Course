```python
#enumerate позволяет получить индекс каждого элемента 
print(list(enumerate([1, 2, 3, 10, 18, 40])))
for i in list(enumerate([1, 2, 3, 10, 18, 40])):
    print(i)
```

```python
companies_capitalization = [
    ['Orange', 1.3],
    ['Maxisoft', 1.5], 
    ['Headbook', 0.8],
    ['Nicola', 2.2]
]
list(enumerate(companies_capitalization))
for i, company in enumerate(companies_capitalization):
    print(i+1, company[0], 'capitalization is', company[1])

for i, _ in enumerate(companies_capitalization):
    print(i+1)    # повторение. нижнее подчеркивание - когда мы игнорируем вторую переменную в составе нашей структуры данных, но должны выполнить соглашение, что на вход приходит две переменных
```

```python
for i, company in enumerate(companies_capitalization):
    print(company)
    if i > 1:
        break
    
print("--------")

for i, company in enumerate(companies_capitalization):
    if i > 1:
        break
    print(company)    # разница в выводе в зависимости от того, в каком месте мы ставим if .. break.
```

List comprehension

```python
# Дана последовательность чисел. Мы хотим оставить только те, которые делятся на 5. 
sequence = range(0, 40, 3)
list(sequence)
```

```python
# Решение в лоб:
for num in sequence:
    if num % 5 == 0:
        print(num)
```

```python
# Решение c записью в список:
filtered_sequence = [] # Сначала создаем этот список
for num in sequence:
    if num % 5 == 0:
        filtered_sequence.append(num)
print(filtered_sequence)
```

```python
# Но это будет сделать короче с помощью List Comprehension. Синтаксис будет таков:
print([num for num in sequence if num % 5 == 0])
# или с действием умножения или другим действием. Условие пишем справа от цикла. Полезное действие - слева от "for".
print([num * 2 for num in sequence if num % 5 == 0])
```

Множества

```python
data_scientist_skills = set(['Python', 'R', 'SQL', 'Tableau', 'SAS', 'Git'])
data_engineer_skills = set(['Python', 'Java', 'Scala', 'Git', 'SQL', 'Hadoop'])
print(data_scientist_skills.union(data_engineer_skills))  # то же самое можно записать в следующей форме, объединения множеств (при этом удалятся дубликаты, тк это свойство множеств), 
# то есть что знает дата инженер либо дата саэнтист:
print(data_scientist_skills | data_engineer_skills)
# это был пример логического "или". Ниже рассмотрим пример логического "и"
```

```python
print(data_scientist_skills.intersection(data_engineer_skills))
print(data_scientist_skills & data_engineer_skills) # = логическое "и" (что знает и дата саэнтист и дата инженер)
```

```python
# Разность множеств: Что знает дата саэнтист, но не знает дата инженер, и наоборот:
print(data_scientist_skills.difference(data_engineer_skills))
print(data_scientist_skills - data_engineer_skills)
print()
# А теперь наоборот:
print(data_engineer_skills.difference(data_scientist_skills))
print(data_engineer_skills - data_scientist_skills)
```
