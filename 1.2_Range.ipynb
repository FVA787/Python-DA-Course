```python
#enumerate позволяет получить индекс каждого элемента 
print(list(enumerate([1, 2, 3, 10, 18, 40])))
for i in list(enumerate([1, 2, 3, 10, 18, 40])):
    print(i)
```

```python
companies_capitalization = [
    ['Orange', 1.3],
    ['Maxisoft', 1.5], 
    ['Headbook', 0.8],
    ['Nicola', 2.2]
]
list(enumerate(companies_capitalization))
for i, company in enumerate(companies_capitalization):
    print(i+1, company[0], 'capitalization is', company[1])

for i, _ in enumerate(companies_capitalization):
    print(i+1)    # повторение. нижнее подчеркивание - когда мы игнорируем вторую переменную в составе нашей структуры данных, но должны выполнить соглашение, что на вход приходит две переменных
```

```python
for i, company in enumerate(companies_capitalization):
    print(company)
    if i > 1:
        break
    
print("--------")

for i, company in enumerate(companies_capitalization):
    if i > 1:
        break
    print(company)    # разница в выводе в зависимости от того, в каком месте мы ставим if .. break.
```
