List comprehension

```python
# Дана последовательность чисел. Мы хотим оставить только те, которые делятся на 5. 
sequence = range(0, 40, 3)
list(sequence)
```

```python
# Решение в лоб:
for num in sequence:
    if num % 5 == 0:
        print(num)
```

```python
# Решение c записью в список:
filtered_sequence = [] # Сначала создаем этот список
for num in sequence:
    if num % 5 == 0:
        filtered_sequence.append(num)
print(filtered_sequence)
```

```python
# Но это будет сделать короче с помощью List Comprehension. Синтаксис будет таков:
print([num for num in sequence if num % 5 == 0])
# или с действием умножения или другим действием. Условие пишем справа от цикла. Полезное действие - слева от "for".
print([num * 2 for num in sequence if num % 5 == 0])
```

Множества

```python
data_scientist_skills = set(['Python', 'R', 'SQL', 'Tableau', 'SAS', 'Git'])
data_engineer_skills = set(['Python', 'Java', 'Scala', 'Git', 'SQL', 'Hadoop'])
print(data_scientist_skills.union(data_engineer_skills))  # то же самое можно записать в следующей форме, объединения множеств (при этом удалятся дубликаты, тк это свойство множеств), 
# то есть что знает дата инженер либо дата саэнтист:
print(data_scientist_skills | data_engineer_skills)
# это был пример логического "или". Ниже рассмотрим пример логического "и"
```

```python
print(data_scientist_skills.intersection(data_engineer_skills))
print(data_scientist_skills & data_engineer_skills) # = логическое "и" (что знает и дата саэнтист и дата инженер)
```

```python
# Разность множеств: Что знает дата саэнтист, но не знает дата инженер, и наоборот:
print(data_scientist_skills.difference(data_engineer_skills))
print(data_scientist_skills - data_engineer_skills)
print()
# А теперь наоборот:
print(data_engineer_skills.difference(data_scientist_skills))
print(data_engineer_skills - data_scientist_skills)
```

```python
# Симметричная разность множеств: Что такого знают дата саэнтист и дата инженер, чего не знают они оба (то есть обратная операция логическому "и"):
print(data_scientist_skills.symmetric_difference(data_engineer_skills))
print(data_engineer_skills ^ data_scientist_skills)
"SQL" in data_scientist_skills
```

```python
# В теории, чтобы избежать ошибки, мы должны делать проверку, прежде чем увеличивать значение:
salaries = {
    'John': 1200 ,
    'Mary': 500,
    'Steven': 1000,
    'Liza': 1500,
    'Andrey': 9000
}

recruit = 'Amanda'
if recruit in salaries:
    print('Значение для ключа уже существует')
else:
    print('Добавляю элемент в словарь')
    salaries[recruit] = 2200
print(salaries)
```

Чтобы такие проверки упростить, существует метод setdefault, который умеет делать простую операцию: он проверяет есть ли в словаре ключ и, если его нет, он добавляет к нему значение, стоящее справа. Если ключ уже есть, он не перезатирает данные, значение этого ключа останется неизменным. Но если ключа нет, он добавит его и установит значение по умолчанию, стоящее справа после запятой в синтаксисе:


```python
salaries = {
    'John': 1200 ,
    'Mary': 500,
    'Steven': 1000,
    'Liza': 1500,
    'Andrey': 9000,
    'Amanda': 2200
}
salaries.setdefault('Paul', 2300)
salaries
```

```python
categories = ['Еда', 'Авто', 'Политика', '215456']
audience = [100, 200, 300]
categories_dict = dict(zip(categories, audience))
print(categories_dict)

# Вот как он работает пошагово:
# zip(categories, audience): Склеивает списки как «молнию» на куртке. Получаются пары: ('Еда', 100), ('Авто', 200) и ('Политика', 300). 
# Значение '215456' отбрасывается, так как для него нет пары во втором списке.
# # dict(...): Превращает эти пары в полноценный словарь {ключ: значение}. dict — это встроенная функция (имя класса) в Python для создания словарей
# Это встроенный тип данных.
```

```python
results = [('date', '2018-01-01'), ('counter', '777'), ('visits', 154)]
data_dict = {metric: value for (metric, value) in results}  #скобки для наглядности, чтобы сделать код симметричным и понятным, чтобы показать что мы распаковываем именно кортежи.

print(data_dict)

# Этот код использует конструкцию, которая называется Dict Comprehension (словарное включение). Это элегантный и быстрый способ создать словарь из любого итерируемого объекта в одну строку.
# Вот как он работает пошагово:
# 1. Источник данных (results)
# У вас есть список кортежей: [('date', '2018-01-01'), ('counter', '777'), ('visits', 154)]. Каждая пара внутри — это готовые "ключ" и "значение".
# 2. Цикл for (metric, value) in results
# Программа пробегает по списку и на каждом шаге делает распаковку кортежа:
# Первый элемент кортежа (например, 'date') попадает в переменную metric.
# Второй элемент (например, '2018-01-01') попадает в переменную value.
# 3. Формирование пары {metric: value}
# Внутри фигурных скобок { } Python берет эти переменные и вставляет их в новый словарь. Двоеточие указывает, что metric станет ключом, а value — его значением.
# Интересный факт:
# Конкретно в этом случае (когда данные уже лежат в виде пар) можно сработать еще быстрее, просто вызвав функцию dict():
# data_dict = dict(results) — результат будет абсолютно таким же! Рассмотрим в ячейке ниже:
```
