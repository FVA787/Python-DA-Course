Вспоминаем один из принципов PEP20: flat is better than nested. Плоское лучше, чем вложенное. Поэтому оптимизируем наш код. Из списка мы можем получить строку. Split - это получение из строки списка. А обратная операция - из списка получить строку - это метод join, он склеивает список со строками в одну большую строку. Мы склеиваем имена по разрыву строки через '\n'.  '\n' - это разрыв строки.

```python
professions = ['IT', 'Искусство', 'Спорт', 'Музыка']
persons = [['Гейтс', 'Джобс', 'Возняк'],['Пикассо', 'Моне', 'Ван Гог'],['Федерер', 'Надаль'],['Моцарт', 'Бах']]
for pro, names in zip(professions, persons):
    print(pro)
    print(names)
    print(*zip(professions, persons))
    print(f'{pro}:')
    print('AA'.join(names))   # /n - это разрыв строки. Метод join в данном случае склеивает элементы списка через разрыв строки
    print(" ItsProfession ".join(professions))
    print()
```

Ниже рассмотрим еще один способ. Этот синтаксис называется List Comprehension (генератор списка). Чтобы его понять, нужно прочитать его не как одну строку, а как «матрешку» из циклов.
В Python генератор списка всегда строится по схеме: результат → откуда берем → условия (если есть)
Этот код делает ровно то же самое, что и вот эти пять строк:
python
flat_list = []
for sublist in my_list:        # Первый for в твоей строке
    for name in sublist:       # Второй for в твоей строке
        flat_list.append(name) 
Но он делает это в одну строку. Если по шагам, то код имеет следующий смысл:
for sublist in my_list: «Возьми каждый вложенный список (например, ['Гейтс', 'Джобс']) из большого списка my_list».
for name in sublist: «Теперь зайди внутрь этого вложенного списка и возьми каждое имя по очереди».
name (самое первое): «Возьми это имя и положи его в мой новый список flat_list».

```python
flat_list = [name for sublist in my_list for name in sublist]  #name — это результат, который мы хотим положить в новый список.
print(flat_list)
```

Напишем игру "угадай число". 

```python
import random
number = random.randint(0, 100)
max_tries = 3
print(number, flush = True)
while max_tries != 0:
    user_number = int(input('введите число от 1 до 100'))
    if user_number > number:
        if max_tries > 1:
            print("Введенное число больше", flush = True)
    elif user_number < number:
        if max_tries > 1:
            print("Введенное число меньше", flush = True)
    else: 
        print("Ты победил!", flush = True)
        break
    max_tries -= 1
else:
    print("Ты проиграллллллл!")
    print(f'Загаданное число было {number}')
```

```python
phone_numbers = 'Мария: 8-743-342-23-32 Александр: 8-323-432-23-67 Андрей +7-845-238-43-44'
phone_pattern = '\d-\d{3}-\d{3}-\d{2}-\d{2} | \+\d-\d{3}-\d{3}-\d{2}-\d{2}'  # \слэш перед плюсом - это экранирование, когда мы ищем сам плюс
#также мы указали два паттерна через вертикальную черту, и можем указать разные паттерны поиска телефонных номеров, написанных по-разному. 
#Если паттерны сложные или имеют общие части, лучше использовать скобки, чтобы «ИЛИ» работало локально:
# r'(abc|def)\d+' — найдет либо abc, либо def, после которых идут цифры.
print(re.findall(phone_pattern, phone_numbers))
```

```python
import re
flight = """
Boarding pass: LA4214 AER-CDB 06NOV
Boarding pass: LA4215 AER-CDB 07NOV
"""
regex_flight = '([A-Z]{2})(\d{4})\s([A-Z]{3})-([A-Z]{3})\s(\d{2}[A-Z]{3})'
flight_match_result = re.search(regex_flight, flight)
print(flight_match_result)
print(flight_match_result.group())
print(flight_match_result.group(1)) #мы можем получить отдельные скобочные группы по их номерам. Номера скобочных групп начинаются с 1.
print(flight_match_result.group(2))
```

findall возвращает список кортежей с совпадениями по скобочным группам:

```python
import re
flight = """
Boarding pass: LA4214 AER-CDB 06NOV
Boarding pass: LA4215 AER-CDB 07NOV
"""
regex_flight = '([A-Z]{2})(\d{4})\s([A-Z]{3})-([A-Z]{3})\s(\d{2}[A-Z]{3})'

results = re.findall(regex_flight, flight)
print(results, '\n')
print(*results, '\n')
print(*results, sep = '\n')
```

```python
messages = ['Опять дождь! Лайков: 2', 'Крутой был концерт 21 августа! Лайков: 28, Ответов: 22', 'Круто поставил лайков 5 6 Лайков: 10, Ответов: 8']
likes = 0
replies = 0
for msg in messages:
    likes_pattern = re.search('Лайков:\s(\d+)', msg)
    replies_pattern = re.search('Ответов:\s(\d+)', msg)
    if likes_pattern:
        likes += int(likes_pattern.group(1))
    if replies_pattern:
        replies += int(replies_pattern.group(1))
print(likes)
print(replies)
```
