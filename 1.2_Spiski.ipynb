```python
#Задача на теорию Коллаца - приведение любого числа к единице путем действий в зависиомости от того, четное оно или нечетное
n = int(input('введите число'))
counter = 1
while n != 1:
    if n % 2 == 0:
        n /= 2
    else:
        n = (n * 3 + 1) / 2
    counter += 1

print(n)
print(counter) 
```

```python
import re
text = 'Python - высокоуровневый язык программирования, ориентированный на всестороннее повышение производительности разработчика и читаемости кода.'
res = []
for i in re.sub('[^\w\s]', '', text).split():    # i - это слово в тексте. Сплитим очищенный от знаков препинания текст. Сплит по умолчанию делит по "пробелу". 
    #Как опциональный аргумент, можно указать любой другой символ в качестве разделителя. Любые пробельные символы в тексте это тоже символы, они считаются. 
    #print(i)
    if len(i) > 10:  #len(str) - возвращает количество символов в строке
        res.append(i)  #прибавляем его к списку [res] в случае, если количество букв более 10
print(res)
```

```python
sum_ = 0

while True:  # Добавляем цикл, чтобы код повторялся. While true - когда мы не знаем конкретного условия выхода, и мы пропишем его искусственно внизу (==0, например)
    number = int(input('введите число: '))
    
    if number == 0:
        break  # Теперь break находится внутри цикла и сработает верно (break работает только внутри while и for)
    else:
        sum_ += number
        print(f"Текущая сумма: {sum_}")
```

```python
professions = ['IT', 'Искусство', 'Спорт', 'Музыка']
persons = [['Гейтс', 'Джобс', 'Возняк'],['Пикассо', 'Моне', 'Ван Гог'],['Федерер', 'Надаль'],['Моцарт', 'Бах']]
```

```python
for pro, names in zip(professions, persons):    #функция zip склеивает элементы именно в кортежи (не в списки). 
    #Сам результат zip — это итератор, но если превратить его в список через list(), мы увидим именно список кортежей.
    print(f'{pro}:')
    for name in names:
        print(f'{name}->> известен за свои достижения в')
    print()
```

```python
print(list(zip(professions, persons))) #через создание списка кортежей (первая строка вывода, мы видим, в квадратных скобках)
print(*zip(professions, persons))   #через указатель на ячейку, так как просто zip выведет нам адрес ячейки, где хранится итератор
```

Вспоминаем один из принципов PEP20: flat is better than nested. Плоское лучше, чем вложенное. Поэтому оптимизируем наш код. Из списка мы можем получить строку. Split - это получение из строки списка. А обратная операция - из списка получить строку - это метод join, он склеивает список со строками в одну большую строку. Мы склеиваем имена по разрыву строки через '\n'.  '\n' - это разрыв строки.

```python
professions = ['IT', 'Искусство', 'Спорт', 'Музыка']
persons = [['Гейтс', 'Джобс', 'Возняк'],['Пикассо', 'Моне', 'Ван Гог'],['Федерер', 'Надаль'],['Моцарт', 'Бах']]
for pro, names in zip(professions, persons):
    print(pro)
    print(names)
    print(*zip(professions, persons))
    print(f'{pro}:')
    print(names)
    print('AAA'.join(names))   # /n - это разрыв строки. Метод join в данном случае склеивает элементы списка через разрыв строки
    print()
```
